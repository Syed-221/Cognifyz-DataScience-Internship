## Task 1: Table Booking and Oniline delivery

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pltb
```

```python
df=pd.read_csv('Dataset .csv')
df.head()
```

```python
print(df.columns)
```

```python
table_booking_counts=df['Has Table booking'].value_counts()
table_booking_percent=df['Has Table booking'].value_counts(normalize=True)*100
print(table_booking_counts)
print(table_booking_percent)
```

```python
online_delivery_counts=df['Has Online delivery'].value_counts()
online_delivery_percent=df['Has Online delivery'].value_counts(normalize=True)*100
print(online_delivery_counts)
print(online_delivery_percent)
```

```python
avg_rating_by_booking=df.groupby('Has Table booking')['Aggregate rating'].mean()
print(avg_rating_by_booking)

sns.barplot(x='Has Table booking',y='Aggregate rating',data=df,palette='Set1')
plt.title('Average Rating by Table Booking')
plt.xlabel('Has Table Booking')
plt.ylabel('Average Rating')
plt.show()
```

```python
print(df['Price range'].value_counts())

price_delivery_counts=df.groupby('Price range')['Has Online delivery'].value_counts()
print(price_delivery_counts)

price_delivery_counts.plot(kind='bar',stacked=True,figsize=(10,6),color=['red','green'])
plt.title('Online delivery Availibility across Price range')
plt.xlabel('Price Range')
plt.ylabel('Number of Resturants')
plt.legend(title='Online Delivery')
plt.show()
```

## Task 2:Price Range Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('Dataset .csv')
df.head()
```

```python
price_counts=df['Price range'].value_counts()
print(price_counts)
most_common_price=price_counts.idxmax()
print(f'The most common price range is {most_common_price}')
```

```python
avg_price_by_rating=df.groupby('Aggregate rating')['Price range'].mean()
print(avg_price_by_rating)
```

```python
if 'Rating color' in df.columns:
    rating_color='Rating color'
elif 'Rating color' in df.columns:
    rating_color='Rating color'
else:
    rating_color=None
if rating_color:
    rating_color_group=df.groupby(['Price range',rating_color])['Aggregate rating'].mean().reset_index()


    best_rating=rating_color_group.loc[rating_color_group['Aggregate rating'].idxmax()]
    print(f"\n The color that represents the best rating for each price range is {best_rating}")
    print(best_rating)
else:
    print("Rating color column not found in the DataFrame.")
```

## Task 3: Feature Engineering


```python
import pandas as pd
df=pd.read_csv('Dataset .csv')
df.head()
```

```python
df['Name Length']=df['Restaurant Name'].astype(str).apply(len)
df['Address length']=df['Address'].astype(str).apply(len)
df['City length']=df['City'].astype(str).apply(len)
df['Locality length']=df['Locality'].astype(str).apply(len)
df[['Address length','Name Length','City length','Locality length']].head()
```

```python
df['Has Table booking']=df['Has Table booking'].map({'Yes':1,'No':0})
df['Has Online delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
df[['Has Table booking','Has Online delivery']]
```
